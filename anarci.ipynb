# Determine sequence distance from closest germline

```
%%capture
%pip install biopython
```

```
from anarci import anarci
from tqdm.auto import tqdm
from pathlib import Path
import numpy as np 
import pandas as pd
import random
```

```
# create path variables
DRIVE = Path()/'drive'/'MyDrive'
MODEL_PATH = DRIVE/'models'
DATA_PATH = DRIVE/'data'
```

```
lstm_scores = pd.read_csv(DATA_PATH/'lstm_scores.csv')
seqs = [('seq'+str(i), seq) for i, seq in enumerate(lstm_scores.seqs.values)]
```

```
# test seq
seq = [('seq1', 'QVQLKESGPGLVAPSQSLSITCTVSGFSLTGYGVNWVRQPPGKGLEWLGMIWGDGSTDYNSALKSRLSISKDNSKSQVFLKMNSLQTDDTARYYCARDGEDYDATFYWYFDVWGAGTTVTVSS')]
```

```
results = anarci(seq, scheme="aho", output=False, assign_germline=True, allowed_species=['human'])
numbering, alignment_details, hit_tables = results
```

```
alignment_details
```

```
def get_alignment_score(alignment):
    scores = alignment[0]['germlines']['v_gene']
    return scores[0][1], scores[1]

alignment_scores = [get_alignment_score(a) for a in alignment_details]
columns = ['gene', 'alignment_score']
alignments_df = pd.DataFrame(alignment_scores, columns=columns)
```

```
scores = pd.concat([lstm_scores, alignments_df], axis=1)
scores.to_csv(DATA_PATH/'scores.csv')
```
